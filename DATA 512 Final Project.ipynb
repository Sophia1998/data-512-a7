# Data Cleaning

```python
df_confirm = pd.DataFrame(us_confirm_data[us_confirm_data['Admin2'] =='Essex'])
df_confirm = pd.DataFrame(df_confirm[us_confirm_data['Province_State'] =='Massachusetts'])
df_confirm = df_confirm.drop(['Province_State', 'Admin2','UID','iso2' ,'iso3', 'code3','FIPS','Country_Region','Lat','Long_','Combined_Key'], axis=1)
```

```python
df_death = pd.DataFrame(us_death_data[us_death_data['Admin2'] =='Essex'])
df_death = pd.DataFrame(df_death[us_death_data['Province_State'] =='Massachusetts'])
df_death = df_death.drop(['Province_State', 'Admin2','UID','iso2' ,'iso3', 'code3','FIPS','Country_Region','Lat','Long_','Combined_Key','Population'], axis=1)
```

```python
df_confirm.head()
```

```python
df_death.head()
```

```python
#Convert string to date
def revertStrToDate(date_str_list):
    datelist = []
    for date_str in date_str_list:
        date_temp = datetime.strptime(date_str, '%m/%d/%y').date()
        datelist.append(date_temp)
    return datelist
```

```python
confirm_cum = df_confirm.values.tolist()[0]
```

```python
daily_cases = []
i = 0
for i in range(len(confirm_cum) - 1):
    con = confirm_cum[i + 1] - confirm_cum[i]
    daily_cases.append(con)
    i = i + 1
daily_cases.append(0)
```

```python
date = list(df_death.columns)
datelist2 = revertStrToDate(date)
death_cum = df_death.values.tolist()[0]
daily_death = []
i = 0
for i in range(len(confirm_cum) - 1):
    con = death_cum[i + 1] - death_cum[i]
    daily_death.append(con)
    i = i + 1
daily_death.append(0)
```

```python
df_covid = pd.DataFrame(datelist2, columns=['date'])
df_covid.insert(loc=1, column='death', value = death_cum)
df_covid.insert(loc=2, column='daily_death', value = daily_death)
df_covid.insert(loc=3, column='confirmed_cases', value = confirm_cum)
df_covid.insert(loc=4, column='daily_new_cases', value = daily_cases)
```

```python
df_covid_factor = df_covid
date_list = df_covid_factor['date'].to_list()
new_d = []
for d in date_list:
    temp = datetime(d.year, d.month, 1)
    new_d.append(temp)
df_covid_factor = df_covid_factor.drop(['date'],axis = 1)
df_covid_factor.insert(loc=0, column='date', value = new_d)
df_covid_factor = df_covid_factor.groupby(by=[df_covid_factor['date']]).sum()
df_covid_factor.head()
```

```python
growing_factor_confirm = []
daily_cases = df_covid_factor['daily_new_cases'].to_list()
i = 0
for i in range(len(daily_cases) - 1):
    if(daily_cases[i] == 0):
        growing_factor_confirm.append(0)
    else:
        growing_factor_confirm.append(daily_cases[i + 1]/daily_cases[i])
    i = i + 1
growing_factor_confirm.append(0)
```

```python
growing_factor_death = []
daily_death = df_covid_factor['daily_death'].to_list()
i = 0
for i in range(len(daily_death) - 1):
    if(daily_death[i] == 0):
        growing_factor_death.append(0)
    else:
        growing_factor_death.append(daily_death[i + 1]/daily_death[i])
    i = i + 1
growing_factor_death.append(0)
```

```python
df_covid_factor.insert(loc=3, column='growing_factor_confirm', value = growing_factor_confirm)
df_covid_factor.insert(loc=4, column='growing_factor_death', value = growing_factor_death)
df_covid_factor.head()
```

```python
n_date = pd.date_range(start="2019-01-01",end="2021-10-01",freq = pd.offsets.MonthBegin(1)).to_pydatetime().tolist()
date = []
for d in n_date:
    date.append(d.date())
```

```python
df_unemployment_master = pd.DataFrame(date, columns=['date'])
df_unemployment_master['unemployment'] = df_unemployment['Value']
df_unemployment_master['labor_force'] = df_labor_force['Value']
df_unemployment_master['unemployment_rate_essex'] = df_unemployment_rate_essex['Value']
df_unemployment_master['unemployment_rate_ma'] = df_unemployment_rate_ma['Value']
```

```python
from datetime import date
d = date(2019, 12, 21)
df_em_2020 = df_unemployment_master
df_em_2020 =  pd.DataFrame(df_em_2020[df_em_2020['date'] > d])
df_unemployment_master = df_unemployment_master.set_index('date')
df_em_2020 = df_em_2020.set_index('date')
df_covid = df_covid.set_index('date')
```

```python
df_master = df_em_2020.join(df_covid_factor, how="inner")
```
